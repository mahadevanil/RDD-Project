```python
#from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("RDD").getOrCreate()
sc= spark.sparkContext # get SparkContext from Spark session
```

```python
df = spark.read.csv("/content/drive/MyDrive/Iris.csv",inferSchema=True,header = True)
df.show()
```

```python
# with & operator
df.filter((df["SepalWidthCm"]<3.5) & (df['PetalWidthCm']>0.2)).show()
```

```python
# with or operator(|)
df.filter((df["SepalWidthCm"]<3.5) | (df['PetalWidthCm']>0.2)).show()
```

```python
#using not operator (~ negation)
df.filter((df["SepalWidthCm"]<3.5) &~ (df['PetalWidthCm']>0.2)).show()
```

```python
df.filter(df["SepalWidthCm"]==3.5).show()
```

```python
#collecting result as python objects
df.filter(df["SepalWidthCm"]==3.5).collect()
```

```python
df.printSchema()
```

```python
df.withColumn('SepalLengthCm_double', df['SepalLengthCm'] * 2).show()
```

```python
#create temperory view
df.createOrReplaceTempView("iris")
```

```python
spark.sql("SELECT * FROM iris LIMIT 10").show()
```

```python
spark.sql("SELECT SepalLengthCm,species From iris").show()
```

```python
spark.sql("SELECT * FROM iris WHERE PetalWidthCm = 0.2").show()
```

```python
spark.sql("SELECT * FROM iris WHERE SepalWidthCm > 3.0  AND Species = 'Iris-setosa'").show()
```

```python
spark.sql("SELECT * FROM iris WHERE SepalWidthCm > 3.0  AND Species = 'Iris-setosa'").show()
```
